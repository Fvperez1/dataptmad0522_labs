```python
# Libraries
import numpy as np
import matplotlib.pyplot as plt
import pandas as pd
import seaborn as sns
import math
sns.set()
```

# Ejemplo de linea

```python

#dibujo de line
plt.figure(figsize=(10, 6))#cambia el tamaño de la figura

#Pais1
x = [2016, 2017, 2018, 2019, 2020, 2021]
y = [42, 43, 45, 47, 48, 50]

#Pais2
x2 = [2016, 2017, 2018, 2019, 2020, 2021]
y2 = [43, 43, 44, 44, 45, 45]


plt.plot(x, y, marker='v', linestyle='--', color='blue', label='Colombia')#datos primer Pais
plt.plot(x2, y2,marker='o', linestyle='-', color='red', label='Argentina')#datos segundo Pais
plt.xlabel('años')#nombre eje x
plt.ylabel('Poblacion en millones')
plt.title('Años vs Poblacion')
plt.legend(loc='lower right')#para que aparezca las leyendas

plt.savefig('ejemplo.png')#guarda el archivo

plt.yticks([41, 45, 48, 51]) #Muestra los numeros que queremos de un eje
plt.show()#muestra solamente el grafico, sin decir que tipo es
```

#  Ejemplo barras

```python
x = ['Argentina', 'Colombia', 'Peru']
y = [40, 50, 33]

plt.bar(x, y)
plt.show()
```

# Piecharts

```python
x = ['Argentina', 'Colombia', 'Peru']
y = [40, 50, 33]

plt.pie(y, labels=x)
plt.show()
```

# Histogram

```python
edades = [15, 16, 17, 20, 21, 22, 23, 24, 25, 26, 30, 31, 32, 35]
bins = [15, 20, 25, 30, 35]
plt.hist(edades, bins, edgecolor='black')#se le añade edgecolor para diferenciar cada barra
plt.show()
```

# BOXPLOT

```python
#Sirve para ver un dato atipico
edades = [15, 16, 17, 20, 21, 22, 23, 24, 25, 26, 30, 31, 32, 35, 70]
plt.boxplot(edades)
plt.show()
```

# Scatterplot

```python
a = [1, 2, 3, 4, 5, 4, 3, 2, 2, 4, 5, 6, 7]
b = [7, 2, 3, 5, 5, 7, 3, 2, 1, 4, 6, 3, 2]
plt.scatter(a, b)
plt.show()
```

#  Subplots

```python
fig, ax = plt.subplots(1, 2, sharey=True)#Se pone Sharey para que compartan el eje y en los dos graficos

ax[0].plot(x, y, color='g', label='Colombia')
ax[0].legend()
ax[1].plot(x2,y2, color='r', label='Argentina')
ax[1].legend()
plt.show()
```
